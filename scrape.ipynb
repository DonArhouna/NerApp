```python
from selenium import webdriver

# Set options to run Chrome in headless mode (no GUI)
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)
```

```python
url_taggat='https://www.defuwaxu.com/xibaari-taggat-yaram-1-10-2023/'
driver.get(url_taggat)
```

```python
page_taggat= driver.page_source
```

```python
page_taggat
```

```python
Art_Taggat=[]
Taggats=driver.find_elements(By.TAG_NAME,'h2')
for Taggat in Taggats:
  if not Taggat.find_elements(By.TAG_NAME,'strong'):
    print(Taggat.text)
    Art_Taggat.append(Taggat.text)
```

```python
# importation des rubriques
Rubrique_Taggat="tdb-entry-category"
```

```python
Rubri_Tag=driver.find_elements(By.CLASS_NAME,Rubrique_Taggat)
```

```python
Rub_Tag=[]
for _ in range(len(Art_Taggat)):
  for Rub in Rubri_Tag:
    Rub_Tag.append(Rub.text)
```
